## <strong><center> Analisando fatores que contribuem<br>para o Diagnóstico de Doenças Cardíacas </center></strong>

![insuficiencia cardiaca.jpg](<attachment:insuficiencia cardiaca.jpg>)

As `doenças cardiovasculares` são a principal causa de morte no mundo, contabilizando cerca de `17,9 milhões de mortes por ano`, o que corresponde a 31% de todas as mortes globais. A maior parte dessas mortes, quatro em cada cinco, é causada por **ataques cardíacos e derrames**, e um terço ocorre prematuramente em pessoas com menos de 70 anos. A `insuficiência cardíaca` é comum nessas condições, e há 11 características neste conjunto de dados que podem auxiliar na previsão de doenças cardíacas.

Indivíduos com doenças cardíacas ou alto risco cardiovascular, devido a fatores como `hipertensão`, `diabetes`, `hiperlipidemia` ou `histórico familiar`, necessitam de diagnóstico e tratamento precoces, nos quais um modelo de aprendizado de máquina pode ser extremamente útil.

## <strong><center> Informações do conjunto de dados </center></strong>

Este conjunto de dados contém `registros relacionados ao diagnóstico de doenças cardíacas em pacientes`. Ele inclui vários atributos clínicos e não clínicos usados ​​para determinar a presença ou ausência de doenças cardíacas. Os dados foram coletados do Kaggle e consistem em 918 linhas com 11 colunas. As colunas abrangem aspectos como dados demográficos do paciente, sinais vitais, sintomas e fatores de risco potenciais associados a condições cardíacas. Este conjunto de dados pode ser utilizado para **análise exploratória** de dados, **modelagem preditiva** e **descoberta de padrões** que contribuem para o diagnóstico de doenças cardíacas.

1 - `Age`: idade do paciente [anos]\
2 - `Sex`: sexo do paciente [M: Masculino, F: Feminino]\
3 - `ChestPainType`: tipo de dor torácica [TA: Angina típica, ATA: Angina atípica, NAP: Dor não anginosa, ASY: Assintomática]\
4 - `ResgingBP`: pressão arterial em repouso [mm Hg]\
5 - `Cholesterol`: colesterol sérico [mm/dl]\
6 - `FastingBS`: glicemia em jejum [1: se BS em jejum > 120 mg/dl, 0: caso contrário]\
7 - `ResgingECG`: resultados do eletrocardiograma em repouso [Normal: Normal, ST: com anormalidade das ondas ST-T (inversões das ondas T e/ou elevação ou depressão de ST > 0,05 mV), HVE: mostrando hipertrofia ventricular esquerda provável ou definitiva pelos critérios de Estes]\
8 - `MaxHR`: frequência cardíaca máxima alcançada [valor numérico entre 60 e 202]\
9 - `ExercícioAngina`: angina induzida por exercício [Y: Sim, N: Não]\
10 - `Oldpeak`: oldpeak = ST [Valor numérico medido na depressão]\
11 - `ST_Slope`: a inclinação do segmento ST do pico do exercício [Up: subida, Flat: flat, Down: downsloping]\
12 - `HeartDisease`: classe de saída [1: doença cardíaca, 0: Normal]\

</br>
Fonte dos dados: https://www.kaggle.com/datasets/amirmahdiabbootalebi/heart-disease

### <center><strong> Visualizando a distribuição dos pacientes por sexo usando o método `value_counts` </strong></center>

No código a seguir, está sendo utilizada a **Biblioteca** *Matplotlib* para criar uma visualização em forma de gráfico de pizza da distribuição de pacientes por sexo em um *DataFrame* chamado **df**. Primeiramente, a **Função** *plt.figure* define o tamanho da figura. Em seguida, a **Função** *value_counts* da coluna '`Sex`' do **DataFrame** **df** conta as ocorrências de cada sexo, e a **Função** *plot.pie* cria o gráfico de pizza com os valores calculados, formatando as porcentagens com duas casas decimais através do **Parâmetro** *autopct*. O título do gráfico é definido pela **Função** *plt.title*. A **Função** *plt.tight_layout* ajusta automaticamente os parâmetros da figura para que a disposição seja mais organizada. Por fim, *plt.show* exibe o gráfico gerado.

```python
plt.figure(figsize=(8,8))
df['Sex'].value_counts().plot.pie(autopct='%1.2f%%',figsize=(8,8))

plt.title("Distribution of Patients by Sex", fontsize=14)
plt.legend(["Masculine", "Feminine"], fontsize=12)
plt.ylabel("Sex", fontsize=12)
plt.tight_layout()

plt.show()
```

Primeiramente, utiliza-se a **Função** `plt.subplots()` da **Biblioteca** `matplotlib` para criar uma figura e um conjunto de eixos com o tamanho especificado pelo **Parâmetro** `figsize`. O título e os rótulos dos eixos são definidos com `plt.title()`, `ax.set_xlabel()` e `ax.set_ylabel()`, ajustando também o tamanho da fonte com o parâmetro `fontsize`. Finalmente, a **Função** `ax.legend()` adiciona uma legenda ao gráfico, e `plt.show()` exibe o gráfico completo.

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.kdeplot(df[df["HeartDisease"] == 1]["Age"], alpha=0.5, fill=True, color="red" , label="Heart Disease", ax=ax)
sns.kdeplot(df[df["HeartDisease"] == 0]["Age"], alpha=0.5, fill=True, color="orange", label="Normal", ax=ax)

plt.title('Distribution of Heart Disease by Patient Age', fontsize=18) 
ax.set_xlabel("Age", fontsize=12) 
ax.set_ylabel("Frequency", fontsize=12)
ax.legend(fontsize=12) 

plt.show()
```

## <center><strong><font color=orange> 8º Analisar possíveis outliers </font></strong></center>

### <center><strong> Carregando as variaveis para `plot` </strong></center>

No código a seguir, está sendo **criada uma lista** chamada **variaveis_numericas** que irá armazenar **nomes de colunas** do **DataFrame** chamado **df** que possuem **tipos de dados** do tipo **int64** ou **float64**. O código percorre as primeiras 11 colunas do **DataFrame** **df** utilizando um **loop for** e, para cada coluna, verifica o seu **tipo de dado**. Se o tipo de dado for **int64** ou **float64**, o **nome da coluna** é adicionado à **lista variaveis_numericas**. Por fim, a **função display** é usada para mostrar o conteúdo da **lista variaveis_numericas**.

```python
variaveis_numericas = []
for i in df.columns[0:11].tolist():
        if df.dtypes[i] == 'int64' or df.dtypes[i] == 'float64':                        
            variaveis_numericas.append(i)   
            
display(variaveis_numericas)
```

No código a seguir, está sendo utilizada a **função** `len` para obter o comprimento da **lista** `variaveis_numericas`, que representa o número total de elementos presentes nessa lista.

```python
len(variaveis_numericas)
```

### <center><strong> Configurando e gerando os boxplots com as funções `subplots` e `boxplot` </strong></center>

No código a seguir, está sendo utilizado a **Biblioteca** `matplotlib` para configurar o tamanho da figura e o layout automático, e a **Biblioteca** `seaborn` para gerar os boxplots. A **Função** `plt.subplots` cria uma grade de subplots com 3 linhas e 2 colunas. Em seguida, um loop `for` percorre as variáveis numéricas de interesse (`variaveis_numericas`), gerando um boxplot para cada variável, utilizando a **Função** `sns.boxplot`, onde o parâmetro `data` define o conjunto de dados (`df`) e o parâmetro `y` define a variável a ser plotada. Os gráficos são posicionados na grade de subplots de acordo com os índices `linha` e `coluna`, que são incrementados para ajustar corretamente a posição dos gráficos. Ao atingir duas colunas, a contagem de colunas é resetada e a contagem de linhas é incrementada. Finalmente, a **Função** `plt.show` exibe todos os gráficos gerados. Com isso, podemos observar nos boxplots abaixo que as variáveis númericas apresentam uma grande quantidade de "possíveis" outliers.

```python
plt.rcParams["figure.figsize"] = [14.00, 20.00]
plt.rcParams["figure.autolayout"] = True

f, axes = plt.subplots(3, 2)

linha = 0
coluna = 0
for i in variaveis_numericas:
    sns.boxplot(data = df, y=i, ax=axes[linha][coluna])
    coluna += 1
    if coluna == 2:
        linha += 1
        coluna = 0            

plt.show()
```

### <center><strong> Criando Gráfico de Mapa de Calor de Correlação com a Função `heatmap()` </strong></center>

No código a seguir, está sendo calculada a **correlação** entre colunas numéricas de um **DataFrame** usando a **função** `corr` da **biblioteca** `pandas`, com o parâmetro `numeric_only=True`. Em seguida, a **biblioteca** `matplotlib` é utilizada para criar uma **figura** de tamanho 12x12, e a **biblioteca** `seaborn` é usada para gerar um **heatmap** da matriz de correlação. O **heatmap** mostra as correlações com anotações e utiliza a **colormap** 'coolwarm', com valores mínimos e máximos definidos entre -1 e 1, centralizado em 0, em um formato quadrado com espessura de linhas de 0.5 e formato numérico com duas casas decimais. O gráfico recebe o título 'Mapa de Calor da Matriz de Correlação' com tamanho de fonte 16 e espaçamento adicional de 20. Em seguida, o código utiliza a biblioteca **"Matplotlib"** e a biblioteca **"Seaborn"** para visualizar essa matriz de covariância. Primeiro, é criada uma figura com dimensões 12x12 usando a função **"figure"** da classe **"plt"** (parâmetro: **"figsize"**). Depois, a função **"heatmap"** da biblioteca **"Seaborn"** é usada para criar um mapa de calor da matriz de covariância (**"cov"**), com vários parâmetros ajustados, incluindo **"annot=True"**, **"cmap='coolwarm'"**, **"vmin=-1"**, **"vmax=1"**, **"center=0"**, **"square=True"**, **"linewidths=0.5"**, e **"fmt='.2f'"**.
